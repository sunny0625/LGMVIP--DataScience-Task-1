```python
def Detect_Fix_Outliers_IQR(df, col_name):
    q3, q1 = df[col_name].quantile(0.75), df[col_name].quantile(0.25)
    high_threshold, low_threshold = q3 + 1.5*(q3 - q1), q1 - 1.5*(q3 - q1)
    
    outlier_booleans = np.logical_or(df[col_name].values > high_threshold, df[col_name].values < low_threshold)
    outlier_indexes = np.where(outlier_booleans)[0]
    
    print(f" Number of Outlier Detected : {len(outlier_indexes)}")
    
    df[col_name] = np.clip(df[col_name], low_threshold, high_threshold)
    
    return df
```

```python
Detect_Fix_Outliers_IQR(df, col_name = 'Sepal_Width(cm)')
```

## Define X & y

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
y.shape, X.shape
```

# Train Test Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 22)
```

```python
X_train
```

```python
print(X_test.shape)
print(y_test.shape)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

## KNeighborsClassifier Report

```python
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy: ' + str(round(accuracy, 2)) + '%')
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 150)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

## LogisticRegression Report

```python
lr_accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy: ' + str(round(lr_accuracy, 2)) + '%')
```

```python
print(classification_report(y_test, y_pred))
```
